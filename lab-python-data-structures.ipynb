#### 3. Ask the user to input the quantity of each product available in the inventory. 
Use the product names from the `products` list as keys in the `inventory` dictionary and assign the respective quantities as values.

````python
#3. user inputs the quantity of each product available in the inventory:
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}
print("Before updating the inventory:", inventory)
print("\nProducts quantities input:")

for product in products:
    quantity = int(input(f"Enter the quantity for {product}: "))
    inventory[product] = quantity
    print(f"- {product} quantity updated to {quantity}.")
print("\nUpdated Inventory:", inventory)
````

##### Error handling the input

````python
# 3. input of the user for quantity of each product
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}
print(inventory)

for i in range(len(products)):
    while True:  # keep asking until valid
        quantity = input("Enter the quantity for " + products[i] + ": ")
        if quantity.isdigit():           # checks if input is a positive integer
            quantity = int(quantity)
            inventory[products[i]] = quantity
            break                        # valid → exit loop
        else:
            print(f"❌ Invalid input! Please enter a numeric quantity for {products[i]}.")

# 3.5 checking our dictionary inventory
print(inventory)
````

#### 5. Ask the user to input the name of three products that a customer wants to order (from those in the products list, meaning three products out of "t-shirt", "mug", "hat", "book" or "keychain". Add each product name to the `customer_orders` set.

````python
# 5. Ask the user to input 3 product names to add to a set called `customers_orders`.----------------
# define list products
products = ["t-shirt", "mug", "hat", "book", "keychain"]
print("Here are the available products :", products)

# Empty customer orders set for testing
customer_orders = set()
print("\n Empty Customer's orders:", customer_orders)

# INPUT 3 PRODUCTS using loop
for i in range(3):

    product = input("Enter product name: ").lower()  # convert input to lowercase
    if product in products:
        print(f"- {product} has been added to your order!")
        customer_orders.add(product)
    else:
        print("Error! Please input product from the list: ", products)
        continue # skip to next iteration if product not in list
    

# updated customer orders set with products for testing
print("\nUpdated Customer's orders:", customer_orders)
````

````python
# 5. using while instead of for loop---------------------------------------------------------------------------------
# unlimited adding of products until We enter 'BUY'
# define list products
products = ["t-shirt", "mug", "hat", "book", "keychain"]
print("Here are the available products :", products)

# Empty customer orders set for testing
customer_orders = set()
print("\n Empty Customer's orders:", customer_orders, "\n")

# must initialize product variable before while loop
product = ""
# INPUT 3 PRODUCTS using loop
while product !="buy": # keep asking for product input until we write buy

    product = input("Enter product name or BUY to complete your purchase: ").lower()  # convert input to lowercase
    if product in products:
        print(f"- {product} has been added to your order!")
        customer_orders.add(product)
    
    elif product == "buy":
        print("Thank you for your purchase!")
        break

    else:
        print(f"Error! {product} is not listed in our products list. Please input product from the list: {products}")
        continue # skip to next iteration if product not in list
    

# updated customer orders set with products for testing
print("\nUpdated Customer's orders:", customer_orders)
````

````python
# 5. using while instead of for loop---------------------------------------------------------------------------------
# limited adding of only 3 products until We enter 'BUY'
# define list products
products = ["t-shirt", "mug", "hat", "book", "keychain"]
print("Here are the available products :", products)

# Empty customer orders set for testing
customer_orders = set()
print("\n Empty Customer's orders:", customer_orders, "\n")

# must initialize product variable before while loop
product = ""
product_nb = 0
# INPUT 3 PRODUCTS using loop
while product !="buy" and product_nb<3: # keep asking for product input until we write buy

    product = input("Enter product name or BUY to complete your purchase: ").lower()  # convert input to lowercase
    if product in products:
        print(f"- {product} has been added to your order!")
        customer_orders.add(product)
        product_nb+=1
    
    elif product == "buy":
        print("Thank you for your purchase!")
        break

    else:
        print(f"Error! {product} is not listed in our products list. Please input product from the list: {products}")
        continue # skip to next iteration if product not in list
    

# updated customer orders set with products for testing
print("\nUpdated Customer's orders:", customer_orders)
````

#### 7. Calculate the following order statistics:
   - Total Products Ordered: The total number of products in the `customer_orders` set.
   - Percentage of Products Ordered: The percentage of products ordered compared to the total available products.
   
   Store these statistics in a tuple called `order_status`.

````python
# 7. Calculate the following order statistics: ------------------------------------------------------------
# 7.1. Total products ordered

total_products_ordered = len(customer_orders)
print(f"\nTotal Products Ordered: {total_products_ordered}")

# 7.2. Percentage products ordered from total products ordered
percentage = total_products_ordered / len(products) * 100

print(f"Percentage of Products Ordered: {percentage:.2f}%")

# Store these statistics in a tuple called `order_status`.

order_status = (total_products_ordered, percentage)
print("Order Status (Total Products Ordered, Percentage):", order_status)
````

#### 9. Update the inventory by subtracting 1 from the quantity of each product. Modify the `inventory` dictionary accordingly.

````python
# 9. Update the inventory dict by substracting 1 from products in customer_orders set ---------------------------------------------
for product in customer_orders:
    if inventory[product] > 0:
        inventory[product] -= 1
        print(f"- {product} inventory decreased by 1. New quantity: {inventory[product]}")
    else:
        print(f"- {product} is out of stock!")
````

-----
#### 10. Print the updated inventory, displaying the quantity of each product on separate lines.

````python
# Print the updated inventory

# print way 1: call dict
print("\nUpdated Inventory after Customer's orders:", inventory)

# print way 2: bullet points using key[i]
print("\nUpdated Inventory after Customer's orders:")
for product in products:
    print(f"- {product}: {inventory[product]} units left")

# print way 3: using items()
print("\nUpdated Inventory after Customer's orders:")
for product, quantity in inventory.items():
    print(f"- {product}: {quantity} units left")

# for key, value in dictionary.items():
#     print(f"{key}: {value}")
````
