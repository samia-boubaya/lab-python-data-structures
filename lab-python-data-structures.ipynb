# Solved Lab 01 | Data Structures

1. Define a list called `products` that contains the following items: "t-shirt", "mug", "hat", "book", "keychain"

````python
# 1.defining the list products
products = ["t-shirt", "mug", "hat", "book", "keychain"]
print(products)
````

2. Create an empty dictionary called `inventory`.

````python
# 2. creating an empty dictionary invetory
inventory = {}
print(inventory)
````

3. Ask the user to input the quantity of each product available in the inventory. Use the product names from the `products` list as keys in the `inventory` dictionary and assign the respective quantities as values.

````python
# 3. input of the user for quantity of each product
for i in range(len(products)):
    while True:  # keep asking until valid
        quantity = input("Enter the quantity for " + products[i] + ": ")
        if quantity.isdigit():           # checks if input is a positive integer
            quantity = int(quantity)
            inventory[products[i]] = quantity
            break                        # valid → exit loop
        else:
            print(f"❌ Invalid input! Please enter a numeric quantity for {products[i]}.")

# 3.5 checking our dictionary inventory
print(inventory)
````

4. Create an empty set called `customer_orders`.

````python
# 4. creating an empty set customer orders
customer_orders = set()
print(customer_orders)
````

5. Ask the user to input the name of three products that a customer wants to order (from those in the products list, meaning three products out of "t-shirt", "mug", "hat", "book" or "keychain". Add each product name to the `customer_orders` set.

````python
# 5. User input for 3 products to order
print("Here is the list of products:")
print(products)

print("Enter the names of 3 products to order:")

while len(customer_orders) < 3:
    item = input("Enter a product name: ").lower()

    if item in products:
        customer_orders.add(item)
        print(f"✅ Added '{item}' to customer orders.")
    else:
        print("❌ Invalid product! Please choose from:", products)
````

6. Print the products in the `customer_orders` set.

````python
# 6. Printing the ordered products
print("Customer Orders:", customer_orders)
````

7. Calculate the following order statistics:
   - Total Products Ordered: The total number of products in the `customer_orders` set.
   - Percentage of Products Ordered: The percentage of products ordered compared to the total available products.
   
   Store these statistics in a tuple called `order_status`.

````python
# 7. Calculating order statistics

# Total Products Ordered
total_products_ordered = len(customer_orders)

# Percentage of Products Ordered
percentage_products_ordered = (total_products_ordered / len(products)) * 100

# Calculate Order Status
order_status = (total_products_ordered, percentage_products_ordered)
# Print Order Status
#print("Order Status (Total, Percentage):", order_status)
````

8. Print the order statistics using the following format:
   ```
   Order Statistics:
   Total Products Ordered: <total_products_ordered>
   Percentage of Products Ordered: <percentage_ordered>% 
   ```

````python
# 8. Storing the statistics in a tuple order_status
print("Order Statistics:")
print("Total Products Ordered:", total_products_ordered)
print(f"Percentage of Products Ordered: {int(percentage_products_ordered)}%")
````

9. Update the inventory by subtracting 1 from the quantity of each product. Modify the `inventory` dictionary accordingly.

````python
# 9. Updating the inventory by subtracting 1 for each ordered product
for ordered_product in customer_orders:
    if inventory[ordered_product] > 0:
        inventory[ordered_product] = inventory[ordered_product] - 1
````

10. Print the updated inventory, displaying the quantity of each product on separate lines.

````python
# 10. Printing the updated inventory
print("Updated Inventory :")
for product, quantity in inventory.items():
    print(product, ":", quantity)
````
